```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, roc_auc_score, confusion_matrix,
    precision_recall_curve, roc_curve, auc
)
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import (
    InputLayer, Conv1D, BatchNormalization,
    MaxPooling1D, Dropout, Flatten, Dense
)
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Wczytaj i przygotuj dane
df = pd.read_csv("/content/sample_data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
if "customerID" in df.columns:
    df.drop(columns=["customerID"], inplace=True)

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df["TotalCharges"].fillna(df["TotalCharges"].median(), inplace=True)
df["Churn"] = df["Churn"].map({"No": 0, "Yes": 1})

# 2. One-hot encoding cech kategorycznych
cat_cols = df.select_dtypes(include=["object"]).columns.tolist()
df_ohe = pd.get_dummies(df, columns=cat_cols, drop_first=True)

# 3. Przygotowanie macierzy cech i wektora etykiet
X = df_ohe.drop(columns=["Churn"]).values.astype(np.float32)
y = df_ohe["Churn"].values.astype(np.int32)

# 4. Standaryzacja cech numerycznych
num_cols = df.select_dtypes(include=["number"]).columns.drop("Churn").tolist()
scaler = StandardScaler()
X[:, :len(num_cols)] = scaler.fit_transform(X[:, :len(num_cols)])

# 5. Podział na train/test i reshape do (samples, features, 1)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
n_features = X_train.shape[1]
X_train = X_train.reshape(-1, n_features, 1)
X_test  = X_test.reshape(-1, n_features, 1)

# 6. Funkcja budująca model 1D-CNN
def build_cnn(input_shape):
    model = Sequential([
        InputLayer(input_shape=input_shape),
        Conv1D(32, 3, activation="relu", padding="same"),
        BatchNormalization(),
        MaxPooling1D(2),
        Dropout(0.2),

        Conv1D(64, 3, activation="relu", padding="same"),
        BatchNormalization(),
        MaxPooling1D(2),
        Dropout(0.3),

        Flatten(),
        Dense(64, activation="relu"),
        Dropout(0.4),
        Dense(1, activation="sigmoid")
    ])
    model.compile(
        loss="binary_crossentropy",
        optimizer=tf.keras.optimizers.Adam(1e-3),
        metrics=["accuracy", tf.keras.metrics.AUC(name="auc")]
    )
    return model

# 7. Utwórz i trenuj model na zbiorze treningowym
cnn = build_cnn((n_features, 1))
history = cnn.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=30,
    batch_size=64,
    verbose=2
)

# 8. Ewaluacja na zbiorze testowym
y_proba = cnn.predict(X_test).ravel()
y_pred  = (y_proba >= 0.5).astype(int)

print("\n=== Test set evaluation ===")
cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm.ravel()
print("Macierz pomyłek (Test):")
print(cm)
print(f"  Prawdziwie Pozytywne (TP) (churn): {tp}")
print(f"  Fałszywie Pozytywne (FP) (churn): {fp}")
print(f"  Prawdziwie Negatywne (TN) (non-churn): {tn}")
print(f"  Fałszywie Negatywne (FN) (churn): {fn}")

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy (Dokładność): {accuracy:.4f}")

recall_churn = tp / (tp + fn) if (tp + fn) > 0 else 0.0
print(f"Recall (Czułość) dla klasy Churn: {recall_churn:.4f}")

precision_churn = tp / (tp + fp) if (tp + fp) > 0 else 0.0
print(f"Precision (Precyzja) dla klasy Churn: {precision_churn:.4f}")

if (precision_churn + recall_churn) > 0:
    f1_churn = 2 * (precision_churn * recall_churn) / (precision_churn + recall_churn)
else:
    f1_churn = 0.0
print(f"F1-score dla klasy Churn: {f1_churn:.4f}")

auroc = roc_auc_score(y_test, y_proba)
print(f"AUROC: {auroc:.4f}")

# 9. Wykresy

# 9.1. Macierz pomyłek (test)
plt.figure(figsize=(4,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Przewidziane')
plt.ylabel('Rzeczywiste')
plt.title('Macierz Pomyłek (zbiór testowy)')
plt.show()

# 9.2. Krzywa ROC
fpr, tpr, _ = roc_curve(y_test, y_proba)
roc_auc_val = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, label=f'AUC = {roc_auc_val:.2f}')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('Odsetek fałszywie pozytywnych')
plt.ylabel('Odsetek prawdziwie pozytywnych (Czułość)')
plt.title('Krzywa ROC')
plt.legend(loc='lower right')
plt.show()

# 9.3. Krzywa uczenia
train_sizes = np.linspace(0.1, 1.0, 5)
train_scores = []
val_scores = []
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
for frac in train_sizes:
    size = int(X_train.shape[0] * frac)
    X_tr_sub = X_train[:size]
    y_tr_sub = y_train[:size]
    fold_train_scores = []
    fold_val_scores = []
    for train_idx, val_idx in skf.split(X_tr_sub, y_tr_sub):
        model = build_cnn((n_features, 1))
        model.fit(X_tr_sub[train_idx], y_tr_sub[train_idx],
                  epochs=10, batch_size=64, verbose=0)
        y_tr_pred = (model.predict(X_tr_sub[train_idx]).ravel() >= 0.5).astype(int)
        y_val_pred = (model.predict(X_tr_sub[val_idx]).ravel() >= 0.5).astype(int)
        fold_train_scores.append(accuracy_score(y_tr_sub[train_idx], y_tr_pred))
        fold_val_scores.append(accuracy_score(y_tr_sub[val_idx], y_val_pred))
    train_scores.append(np.mean(fold_train_scores))
    val_scores.append(np.mean(fold_val_scores))

plt.figure()
plt.plot(train_sizes * X_train.shape[0], train_scores, 'o-', color="r", label='Zbiór treningowy')
plt.plot(train_sizes * X_train.shape[0], val_scores, 'o-', color="g", label='Walidacja krzyżowa')
plt.xlabel('Liczba próbek treningowych')
plt.ylabel('Dokładność (Accuracy)')
plt.title('Krzywa uczenia')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, roc_auc_score, confusion_matrix,
    precision_recall_curve, roc_curve, auc
)
from sklearn.utils import class_weight
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import (
    InputLayer, Conv1D, BatchNormalization,
    MaxPooling1D, Dropout, Flatten, Dense
)
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Wczytaj i przygotuj dane
df = pd.read_csv("/content/sample_data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
if "customerID" in df.columns:
    df.drop(columns=["customerID"], inplace=True)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df["TotalCharges"].fillna(df["TotalCharges"].median(), inplace=True)
df["Churn"] = df["Churn"].map({"No": 0, "Yes": 1})

# 2. One-hot encoding cech kategorycznych
cat_cols = df.select_dtypes(include=["object"]).columns.tolist()
df_ohe = pd.get_dummies(df, columns=cat_cols, drop_first=True)

# 3. Przygotuj macierz cech i wektor etykiet
X = df_ohe.drop(columns=["Churn"]).values.astype(np.float32)
y = df_ohe["Churn"].values.astype(np.int32)

# 4. Standaryzacja cech numerycznych
num_cols = df.select_dtypes(include=["number"]).columns.drop("Churn").tolist()
scaler = StandardScaler()
X[:, :len(num_cols)] = scaler.fit_transform(X[:, :len(num_cols)])

# 5. Split na zbiór treningowy i testowy oraz reshape do (samples, features, 1)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
n_features = X_train.shape[1]
X_train = X_train.reshape(-1, n_features, 1)
X_test  = X_test.reshape(-1, n_features, 1)

# 6.  class_weight
cw = class_weight.compute_class_weight(
    class_weight="balanced",
    classes=np.unique(y_train),
    y=y_train
)
class_weight_dict = {i: w for i, w in enumerate(cw)}

# 7. Funkcja budująca model 1D-CNN
def build_cnn(input_shape):
    model = Sequential([
        InputLayer(input_shape=input_shape),
        Conv1D(32, 3, activation="relu", padding="same"),
        BatchNormalization(),
        MaxPooling1D(2),
        Dropout(0.2),

        Conv1D(64, 3, activation="relu", padding="same"),
        BatchNormalization(),
        MaxPooling1D(2),
        Dropout(0.3),

        Flatten(),
        Dense(64, activation="relu"),
        Dropout(0.4),
        Dense(1, activation="sigmoid")
    ])
    model.compile(
        loss="binary_crossentropy",
        optimizer=tf.keras.optimizers.Adam(1e-3),
        metrics=["accuracy", tf.keras.metrics.AUC(name="auc")]
    )
    return model

# 8. Trenuj model z class_weight
cnn = build_cnn((n_features, 1))
history = cnn.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=30,
    batch_size=64,
    class_weight=class_weight_dict,
    verbose=2
)

# 9. Ewaluacja na zbiorze testowym
y_proba = cnn.predict(X_test).ravel()
y_pred  = (y_proba >= 0.5).astype(int)

print("\n=== Test set evaluation ===")
cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm.ravel()
print("Macierz pomyłek (Test):")
print(cm)
print(f"  Prawdziwie Pozytywne (TP) (churn): {tp}")
print(f"  Fałszywie Pozytywne (FP) (churn): {fp}")
print(f"  Prawdziwie Negatywne (TN) (non-churn): {tn}")
print(f"  Fałszywie Negatywne (FN) (churn): {fn}")

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy (Dokładność): {accuracy:.4f}")

recall_churn = tp / (tp + fn) if (tp + fn) > 0 else 0.0
print(f"Recall (Czułość) dla klasy Churn: {recall_churn:.4f}")

precision_churn = tp / (tp + fp) if (tp + fp) > 0 else 0.0
print(f"Precision (Precyzja) dla klasy Churn: {precision_churn:.4f}")

if (precision_churn + recall_churn) > 0:
    f1_churn = 2 * (precision_churn * recall_churn) / (precision_churn + recall_churn)
else:
    f1_churn = 0.0
print(f"F1-score dla klasy Churn: {f1_churn:.4f}")

auroc = roc_auc_score(y_test, y_proba)
print(f"AUROC: {auroc:.4f}")

# 10. Wykresy

# 10.1. Macierz pomyłek (test)
plt.figure(figsize=(4,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Przewidziane')
plt.ylabel('Rzeczywiste')
plt.title('Macierz Pomyłek (Zbiór Testowy)')
plt.show()

# 10.2. ROC curve
fpr, tpr, _ = roc_curve(y_test, y_proba)
roc_auc_val = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, label=f'AUC = {roc_auc_val:.2f}')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('Odsetek Fałszywie Pozytywnych (False Positive Rate)')
plt.ylabel('Odsetek Prawdziwie Pozytywnych (True Positive Rate)')
plt.title('Krzywa ROC (Receiver Operating Characteristic)')
plt.legend(loc='lower right')
plt.show()


# 10.3. Krzywa uczenia
from sklearn.model_selection import StratifiedKFold

train_sizes = np.linspace(0.1, 1.0, 5)
train_scores = []
val_scores = []
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
for frac in train_sizes:
    size = int(X_train.shape[0] * frac)
    X_tr_sub = X_train[:size]
    y_tr_sub = y_train[:size]
    fold_train_scores = []
    fold_val_scores = []
    for train_idx, val_idx in skf.split(X_tr_sub, y_tr_sub):
        model = build_cnn((n_features, 1))
        model.fit(X_tr_sub[train_idx], y_tr_sub[train_idx],
                  epochs=10, batch_size=64, verbose=0,
                  class_weight=class_weight_dict)
        y_tr_pred = (model.predict(X_tr_sub[train_idx]).ravel() >= 0.5).astype(int)
        y_val_pred = (model.predict(X_tr_sub[val_idx]).ravel() >= 0.5).astype(int)
        fold_train_scores.append(accuracy_score(y_tr_sub[train_idx], y_tr_pred))
        fold_val_scores.append(accuracy_score(y_tr_sub[val_idx], y_val_pred))
    train_scores.append(np.mean(fold_train_scores))
    val_scores.append(np.mean(fold_val_scores))

plt.figure()
plt.plot(train_sizes * X_train.shape[0], train_scores, 'o-', color="r", label='Train')
plt.plot(train_sizes * X_train.shape[0], val_scores, 'o-', color="g", label='CV')
plt.xlabel('Liczba próbek treningowych')
plt.ylabel('Accuracy')
plt.title('Krzywa uczenia (Learning Curve)')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    classification_report, confusion_matrix, accuracy_score,
    roc_auc_score, precision_recall_curve, roc_curve, auc
)
from imblearn.over_sampling import SMOTE
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import (
    InputLayer, Conv1D, BatchNormalization,
    MaxPooling1D, Dropout, Flatten, Dense
)
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Wczytaj i przygotuj dane
df = pd.read_csv("/content/sample_data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
if "customerID" in df.columns:
    df.drop(columns=["customerID"], inplace=True)
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df["TotalCharges"].fillna(df["TotalCharges"].median(), inplace=True)
df["Churn"] = df["Churn"].map({"No": 0, "Yes": 1})

# 2. One-hot encoding cech kategorycznych
cat_cols = df.select_dtypes(include=["object"]).columns.tolist()
df_ohe = pd.get_dummies(df, columns=cat_cols, drop_first=True)

# 3. Przygotuj macierz cech i wektor etykiet
X = df_ohe.drop(columns=["Churn"]).values.astype(np.float32)
y = df_ohe["Churn"].values.astype(np.int32)

# 4. Standaryzacja cech numerycznych
num_cols = df.select_dtypes(include=["number"]).columns.drop("Churn").tolist()
scaler = StandardScaler()
X[:, :len(num_cols)] = scaler.fit_transform(X[:, :len(num_cols)])

# 5. Podział na train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 6. Oversampling SMOTE na zbiorze treningowym
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

# 7. Przygotowanie danych do Conv1D
n_features = X_train_res.shape[1]
X_train_cnn = X_train_res.reshape(-1, n_features, 1)
X_test_cnn  = X_test.reshape(-1, n_features, 1)

# 8. Definicja funkcji budującej model 1D-CNN
def build_cnn(input_shape):
    model = Sequential([
        InputLayer(input_shape=input_shape),
        Conv1D(32, 3, activation="relu", padding="same"),
        BatchNormalization(),
        MaxPooling1D(2),
        Dropout(0.2),

        Conv1D(64, 3, activation="relu", padding="same"),
        BatchNormalization(),
        MaxPooling1D(2),
        Dropout(0.3),

        Flatten(),
        Dense(64, activation="relu"),
        Dropout(0.4),
        Dense(1, activation="sigmoid")
    ])
    model.compile(
        loss="binary_crossentropy",
        optimizer=tf.keras.optimizers.Adam(1e-3),
        metrics=["accuracy", tf.keras.metrics.AUC(name="auc")]
    )
    return model

# 9. Trenowanie modelu na oversamplowanym zbiorze
cnn = build_cnn((n_features, 1))
history = cnn.fit(
    X_train_cnn, y_train_res,
    validation_split=0.1,
    epochs=30,
    batch_size=64,
    verbose=2
)

# 10. Ewaluacja na zbiorze testowym
y_proba = cnn.predict(X_test_cnn).ravel()
y_pred  = (y_proba >= 0.5).astype(int)

print("\n=== Ewaluacja na zbiorze testowym (SMOTE) ===")
cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm.ravel()
print("Macierz pomyłek (Test):")
print(cm)
print(f"  Prawdziwie Pozytywne (TP) (churn): {tp}")
print(f"  Fałszywie Pozytywne (FP) (churn): {fp}")
print(f"  Prawdziwie Negatywne (TN) (non-churn): {tn}")
print(f"  Fałszywie Negatywne (FN) (churn): {fn}")

accuracy = accuracy_score(y_test, y_pred)
print(f"Dokładność (Accuracy): {accuracy:.4f}")

recall_churn = tp / (tp + fn) if (tp + fn) > 0 else 0.0
print(f"Czułość (Recall) dla klasy Churn: {recall_churn:.4f}")

precision_churn = tp / (tp + fp) if (tp + fp) > 0 else 0.0
print(f"Precyzja (Precision) dla klasy Churn: {precision_churn:.4f}")

if (precision_churn + recall_churn) > 0:
    f1_churn = 2 * (precision_churn * recall_churn) / (precision_churn + recall_churn)
else:
    f1_churn = 0.0
print(f"F1-score dla klasy Churn: {f1_churn:.4f}")

auroc = roc_auc_score(y_test, y_proba)
print(f"Pole pod krzywą ROC (AUC): {auroc:.4f}")

# 11. Cross-validation 5-fold ze SMOTE
y_true_cv, y_pred_cv = [], []
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
for train_idx, val_idx in skf.split(X, y):
    # split
    X_tr, y_tr = X[train_idx], y[train_idx]
    X_val, y_val = X[val_idx], y[val_idx]
    # SMOTE oversampling na foldzie
    X_tr_res, y_tr_res = SMOTE(random_state=42).fit_resample(X_tr, y_tr)
    # reshape
    X_tr_cnn = X_tr_res.reshape(-1, n_features, 1)
    X_val_cnn = X_val.reshape(-1, n_features, 1)
    # train
    fold_model = build_cnn((n_features, 1))
    fold_model.fit(X_tr_cnn, y_tr_res, epochs=15, batch_size=64, verbose=0)
    # infer
    y_val_proba = fold_model.predict(X_val_cnn).ravel()
    y_val_pred  = (y_val_proba >= 0.5).astype(int)
    y_true_cv.extend(y_val)
    y_pred_cv.extend(y_val_pred)

cm_cv = confusion_matrix(y_true_cv, y_pred_cv)
acc_cv = accuracy_score(y_true_cv, y_pred_cv)
print("\n=== Walidacja krzyżowa (SMOTE) ===")
print("Macierz pomyłek (CV):\n", cm_cv)
print(f"Dokładność CV: {acc_cv:.4f}")

# 12. Wykresy

# 12.1. Macierz pomyłek (test)
plt.figure(figsize=(4,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Przewidziane')
plt.ylabel('Rzeczywiste')
plt.title('Macierz pomyłek (zbiór testowy)')
plt.show()

# 12.2. Krzywa ROC
fpr, tpr, _ = roc_curve(y_test, y_proba)
roc_auc_val = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, label=f'AUC = {roc_auc_val:.2f}')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('Odsetek fałszywie pozytywnych')
plt.ylabel('Odsetek prawdziwie pozytywnych (Czułość)')
plt.title('Krzywa ROC')
plt.legend(loc='lower right')
plt.show()

# 12.3. Krzywa uczenia
train_sizes = np.linspace(0.1, 1.0, 5)
train_scores = []
val_scores = []
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
for frac in train_sizes:
    size = int(X_train.shape[0] * frac)
    X_tr_sub = X_train[:size]
    y_tr_sub = y_train[:size]
    fold_train_scores = []
    fold_val_scores = []
    for train_idx, val_idx in skf.split(X_tr_sub, y_tr_sub):
        model = build_cnn((n_features, 1))
        model.fit(X_tr_sub[train_idx], y_tr_sub[train_idx],
                  epochs=10, batch_size=64, verbose=0)
        y_tr_pred = (model.predict(X_tr_sub[train_idx]).ravel() >= 0.5).astype(int)
        y_val_pred = (model.predict(X_tr_sub[val_idx]).ravel() >= 0.5).astype(int)
        fold_train_scores.append(accuracy_score(y_tr_sub[train_idx], y_tr_pred))
        fold_val_scores.append(accuracy_score(y_tr_sub[val_idx], y_val_pred))
    train_scores.append(np.mean(fold_train_scores))
    val_scores.append(np.mean(fold_val_scores))

plt.figure()
plt.plot(train_sizes * X_train.shape[0], train_scores, 'o-', color="r", label='Zbiór treningowy')
plt.plot(train_sizes * X_train.shape[0], val_scores, 'o-', color="g", label='Walidacja krzyżowa')
plt.xlabel('Liczba próbek treningowych')
plt.ylabel('Dokładność (Accuracy)')
plt.title('Krzywa uczenia')
plt.legend()
plt.grid()
plt.show()
```
